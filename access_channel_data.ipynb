### Notebook showing how to access MITgcm channel data 

Here I show how to access the data for a channel simulation (ideal ACC) done using MITgcm. 

The data is basically from the 5km run from : https://dhruvbalwada.github.io/assets/documents/Balwada_et_al2018.pdf 

I believe that the SSH can be used as a good first guess for the stream function of the geostrophic flow. It will also be dominantly incompressible. If things work with this, then removing the divergent part (potential) etc can be tried in future iterations. MITgcm is capable of doing this removal, as documented in Appendix A of: https://agupubs.onlinelibrary.wiley.com/doi/full/10.1002/jgrc.20066

```python
# Load some useful modules 
import numpy as np
import xarray as xr
from matplotlib import pyplot as plt
```

```python
#https://catalog.pangeo.io/browse/master/ocean/
from intake import open_catalog

cat = open_catalog("https://raw.githubusercontent.com/pangeo-data/pangeo-datastore/master/intake-catalogs/ocean.yaml")
list(cat) # list the different data sets in the ocean catalog
```

```python
# We will use this https://catalog.pangeo.io/browse/master/ocean/channel/run_tracers_restored_zarr/
#cat = open_catalog("https://raw.githubusercontent.com/pangeo-data/pangeo-datastore/master/intake-catalogs/ocean/channel.yaml")
ds  = cat['channel']['run_tracers_restored_zarr'].to_dask()
```

```python
# this is 3 day output. As there is not much variability at scales smaller than this, there is
# no point saving the output at higher resolution. Also means that intermediate times (not saved)
# can simply be reconstructed by linear interpolation. 
# 
# Datasets at other resolutions are also saved, but not available 
# at high time resolution on the cloud right now.
ds
```

```python
# The variables you care about are probably only u,v and Eta (SSH)
# This data has signatures of geostrophic turbulence and spatial inhomogeneity. 
ds['Eta'].isel(time=0).plot.contourf(levels=15)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(121)
ds['U'].isel(time=0, Z=0).plot.contourf(levels=15)

plt.subplot(122)
ds['V'].isel(time=0, Z=0).plot.contourf(levels=15)
plt.tight_layout()
```

```python
# incase you need to do gradients or somehting
#https://xgcm.readthedocs.io/en/latest/grids.html
import xgcm

grid = xgcm.Grid(ds, periodic='X')
```

```python
# This data is on a cartesian grid with uniform grid spacing. 
# So one can simply do things like 
dx = 5e3
u_x = (grid.diff(ds.U, 'X')/ dx).rename('dudx')
v_y = (grid.diff(ds.V, 'Y', boundary='extend')/ dx).rename('dvdy')

divergence = (u_x+v_y).rename('Div')
```

```python
divergence
```

```python
divergence.isel(time=0, Z=0).plot(vmin=-1e-5)
```

```python
u_y = (grid.diff(ds.U, 'Y', boundary='extend')/ dx).rename('dudy')
v_x = (grid.diff(ds.V, 'X')/ dx).rename('dvdx')

vorticity = (v_x - u_y).rename('vorticity')
```

```python
vorticity # notice that this is on a different x,y grid than divergence 
```

```python
vorticity = grid.interp( grid.interp(vorticity,'X'), 'Y', boundary='extend')
vorticity = vorticity.rename('vorticity')
vorticity
```

```python
vorticity.isel(time=0, Z=0).plot(vmin=-1e-4)
```
